# Plot a bar graph of male vs female buyers participated in the sales

```python
plt.figure(figsize=(8,6))
sns.countplot(x=data['Buyer Gender'])
plt.title('Bar graph of Male VS Female Buyers' , fontsize=15)
plt.grid()
plt.show()
```

# Find the top 5 cars based on their sales price


```python
car5=data.nlargest(5,['Sale Price'])
car5
```

```python
car5_df=pd.DataFrame(car5 , columns=['Make','Model' , 'Sale Price'])
car5_df.iloc[:5]
```

# Find the least 5 cars based on their Resell price

```python
carleast=data.nsmallest(5 , ['Resell Price'])
carleast
```

```python
carleast_df=pd.DataFrame(carleast , columns=['Make','Model' , 'Resell Price'])
carleast_df.iloc[:5]
```
